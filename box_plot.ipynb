```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```python
def box_filter(image, filter_size):
    kernel = np.ones((filter_size, filter_size), dtype=np.float32) / (filter_size * filter_size)
    result = cv2.filter2D(image, -1, kernel)
    return result
```

```python
image = cv2.imread('/mnt/c/Users/DELL/Downloads/pexels-photo-674010.jpeg', cv2.IMREAD_GRAYSCALE)
```

```python
# Apply a 3x3 box filter
filtered_image_3x3 = box_filter(image, 50)

# Apply a 5x5 box filter
filtered_image_5x5 = box_filter(image, 5)
```

```python
# Display the original and filtered images
plt.figure(figsize=(10, 5))
```

```python
plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')
```

```python
plt.subplot(1, 3, 2)
plt.title('Box Filter (3x3)')
plt.imshow(filtered_image_3x3, cmap='gray')
plt.axis('off')
```

```python
plt.subplot(1, 3, 3)
plt.title('Box Filter (5x5)')
plt.imshow(filtered_image_5x5, cmap='gray')
plt.axis('off')
```
