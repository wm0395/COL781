```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft2, fftshift
```

```python
image = plt.imread('/mnt/c/Users/DELL/Downloads/pexels-photo-674010.jpeg')
plt.imshow(image)
```

```python
if len(image.shape) == 3:
    image = np.mean(image, axis=-1)

plt.imshow(image, cmap='grey')
```

```python
frequency_domain = fft2(image)
# print(frequency_domain)

# Shift zero frequency components to the center
frequency_domain_shifted = fftshift(frequency_domain)

# Compute magnitude spectrum (optional)
magnitude_spectrum = np.abs(frequency_domain_shifted)
```

```python
magnitude_spectrum
```

```python
plt.figure(figsize=(12, 6))
```

```python
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')
```

```python
plt.subplot(1, 2, 2)
plt.title('Frequency Domain')
plt.imshow(np.log(1 + magnitude_spectrum), cmap='gray')
plt.axis('off')
```
